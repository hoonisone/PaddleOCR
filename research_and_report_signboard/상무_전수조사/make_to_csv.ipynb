```python
import json
import pandas as pd
path = "/home/resources/cropped_sign_result/system_results.txt"

with open(path) as f:
    lines = [line.rstrip("\n").split("\t") for line in f.readlines()]

a = []
df = pd.DataFrame()
for i, (path, data) in enumerate(lines):
    data = json.loads(data)
    data = [p["transcription"] for p in data]
    df = df.append({"path":path, "keywords":data, "keywords_by_double_tab":"\t\t".join(data)}, ignore_index=True)

df.to_csv("/home/research/상무_전수조사/cropped_sign.csv")
df
```

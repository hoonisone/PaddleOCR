```python
from pathlib import Path
from database import *

def check_missed_weight(work, min_assure, until, interval):
    # 모델 가중치 에폭에 대해 until까지 interval 간격으로 가중치가 필요하다고 할 때 모두 갖춰졌는지 확인
    # 누락, 중복 여부를 판단하여 정리
    miss = []    
    for version in range(1, until+1):
        path = workdb.get_model_weight(work, version, no_exist_handling=False)
        if (version <= min_assure or version%interval == 0):
            if path is None:
                miss.append(version)
    return miss

def check_weight_for_works(works, min_assure, until, interval):
    for work in works:
        missed_versions = check_missed_weight(work, min_assure, until, interval)
        print(work, missed_versions)

workdb = WorkDB()        
works = workdb.get_all_id()
works_10k = [work for work in works if "10k" in work]
works_30k = [work for work in works if "30k" in work]
works_50k = [work for work in works if "50k" in work]
works_full = [work for work in works if "full" in work]


min_assure = 5
check_weight_for_works(works_10k, min_assure, 300, 10)
check_weight_for_works(works_30k, min_assure, 200, 5)
check_weight_for_works(works_50k, min_assure, 100, 5)
check_weight_for_works(works_full, min_assure, 80, 1)
```

Clean Weight Files

```python
from pathlib import Path
from database import *
workdb = WorkDB()
works = workdb.get_all_id()
works_10k = [work for work in works if "10k" in work]
works_30k = [work for work in works if "30k" in work]
works_50k = [work for work in works if "50k" in work]
works_full = [work for work in works if "full" in work]

            

def extract_unnecessary_weights(work, min_assure, interval):
    unnecessary_weights = []
    max_epoch = workdb.get_trained_epoch(work)
    
    for version in range(1, max_epoch+1):
        path = workdb.make_model_weight_path(work, version, relative_to = "absolute", extension="pdparams")
        if (version <= min_assure or version%interval == 0):
            continue
        if Path(path).exists():
                unnecessary_weights.append(path)
                
    for extension in ["pdopt", "states"]:
        for version in range(1, max_epoch+1):
            path = workdb.make_model_weight_path(work, version, relative_to="absolute", extension=extension)
            if Path(path).exists():
                unnecessary_weights.append(path)

    return unnecessary_weights

def delete(path_list):
    for path in path_list:        
        path = Path(path)
        path.unlink()            

def clean_unnecessary_weights(works, min_assure, interval):
    for work in works:
        weights = extract_unnecessary_weights(work, min_assure, interval)
        print(weights)
        delete(weights)
        print(f"deleted {len(weights)}files")

# clean_unnecessary_weights(works_10k, 10, 10)
# clean_unnecessary_weights(works_30k, 10, 5)
# clean_unnecessary_weights(works_50k[:4], 10, 5)
clean_unnecessary_weights(works_full, 10, 1)
```

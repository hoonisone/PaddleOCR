```python
import json

lines = datasetdb.get_all_labels("aihub_rec")
# labels = [json.loads(label.split("\t")[1])["label"] for label in labels]
labels = []
for line in lines:
    try:
        label = json.loads(line.split("\t")[1])["label"]
    except:
        print(line)
        continue
    labels.append(label)

# char_count_dict = {}
char_count_dict = {chr(code):0 for code in range(0xAC00, 0xD7A4)}
for label in labels:
    for c in label:
        if c in char_count_dict:
            char_count_dict[c] += 1
        else:
            char_count_dict[c] = 1

print(char_count_dict)
```

```python

my_dict = char_count_dict

# 값 기준으로 오름차순 정렬
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))

print(sorted_dict)

sum(sorted_dict.values())
```

```python
result1 = "/home/works/rec_PPOCR_full_auto_rotate_v1/inference_result"
result2 = "/home/works/rec_PPOCR_full_auto_rotate_v2/inference_result"
result3 = "/home/works/rec_rotation_pretrained_and_direct_collection_finetuning/inference_result"

with open(result1) as f:
    result1 = [line.rstrip("\n").split("\t") for line in f.readlines()]
    
with open(result2) as f:
    result2 = [line.rstrip("\n").split("\t") for line in f.readlines()]
    
with open(result3) as f:
    result3 = [line.rstrip("\n").split("\t") for line in f.readlines()]
    
for [img1, l1, c1], [img2, l2, c2], [img3, l3, c3] in zip(result1, result2, result3):
    c1, c2, c3 = float(c1), float(c2), float(c3)
    if (c1+0.05 < c2) and (c2+0.05 < c3):
        print(img1, l1, "(%.3f)"%c1, l2, "(%.3f)"%c2, l3, "(%.3f)"%c3,)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 예제 데이터: 2차원 점들의 집합
points = np.array([
    [2, 3], [3, 5], [5, 8], [7, 10], [9, 12],
    [11, 15], [12, 17], [13, 19], [14, 20], [15, 22]
])

# PCA 모델 생성 및 적용 (n_components=2는 전체 주성분을 구하기 위함)
pca = PCA(n_components=2)
pca.fit(points)

# 주성분 추출
principal_axes = pca.components_
mean = pca.mean_

# 데이터 시각화
plt.figure(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], label='Data Points')

# 주성분 축 시각화
for length, vector in zip(pca.explained_variance_, principal_axes):
    v = vector * 3 * np.sqrt(length)  # 주성분 길이를 시각적으로 확대
    plt.plot([mean[0], mean[0] + v[0]], [mean[1], mean[1] + v[1]], 
             label=f'Principal Axis: {vector}', linewidth=3)

plt.title('PCA of 2D Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```
